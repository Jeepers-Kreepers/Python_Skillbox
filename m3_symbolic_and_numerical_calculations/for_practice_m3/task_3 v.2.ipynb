### Задача 3 ###
#### Решите систему уравнений: ####
$$ \left\{\begin{matrix}
\sqrt{x - 4y}-2\sqrt{x + 3y} = 1 \\
7\sqrt{x+3y} - 5x + 22y = 13
\end{matrix}\right. $$

```python
from sympy import symbols, Eq, solve, sqrt

x, y = symbols('x y', real=True)

eq1 = Eq(sqrt(x - 4 * y) - 2 * sqrt(x + 3 * y), 1)
eq2 = Eq(7 * sqrt(x + 3 * y) - 5 * x + 22 * y, 13)

solution = solve((eq1, eq2), (x, y))
print("Символьное решение:", solution)
```

```python
from sympy import symbols, sqrt, plot_implicit

# Определяем переменные
x, y = symbols('x y', real=True)

# Записываем уравнение
equation1 = sqrt(x - 4 * y) - 2 * sqrt(x + 3 * y) - 1
equation2 = 7 * sqrt(x + 3 * y) - 5 * x + 22 * y - 13

# Строим график неявной функции
p1 = plot_implicit(equation1, (x, -10, 10), (y, -10, 10),
                   title="График уравнения: sqrt(x - 4y) - 2*sqrt(x + 3y) = 1",
                   xlabel='x', ylabel='y',
                   line_color='blue',
                   show=False
                   )
p2 = plot_implicit(equation2, (x, -10, 10), (y, -10, 10),
                   title="7*sqrt(x + 3*y) - 5*x + 22*y = 13",
                   xlabel='x', ylabel='y',
                   line_color='red',
                   show=False
                   )

# Объединяем графики
p1.extend(p2)
p1.title = "Графики обоих уравнений системы"
p1.xlabel = 'x'
p1.ylabel = 'y'
p1.legend = True
p1.show()
```

$$ \left\{\begin{matrix}
x^4 - y^4 =  15 \\ x ^ 3 * y - x * y ^ 3  =  6
 \\
\end{matrix}\right. $$

```python
x, y = symbols('x y', real=True)

eq1 = x ** 4 - y ** 4 - 15
eq2 = x ** 3 * y - x * y ** 3 - 6

solution = solve((eq1, eq2), (x, y))
print("Символьное решение:", solution)
```

```python
# Строим график неявной функции
p1 = plot_implicit(eq1, (x, -10, 10), (y, -10, 10),
                   title="График уравнения: x ** 4 - y ** 4 = 15",
                   xlabel='x', ylabel='y',
                   line_color='blue',
                   show=False
                   )
p2 = plot_implicit(eq2, (x, -10, 10), (y, -10, 10),
                   title="x ** 3 * y - x * y ** 3 = 6",
                   xlabel='x', ylabel='y',
                   line_color='red',
                   show=False
                   )

# Объединяем графики
p1.extend(p2)
p1.title = "Графики обоих уравнений системы"
p1.xlabel = 'x'
p1.ylabel = 'y'
p1.legend = True
p1.show()
```

$$ \left\{\begin{matrix}
x - y + z =  2 \\ x ^ 2 + y ^ 2 + z ^ 2 =  6
 \\ x ^ 3 + y ^ 3 + z ^ 3 = 8
\end{matrix}\right. $$

```python
x, y, z = symbols('x y z', real=True)

eq1 = x - y + z - 2
eq2 = x ** 2 + y ** 2 + z ** 2 - 6
eq3 = x ** 3 + y ** 3 + z ** 3 - 8

solutions = solve([eq1, eq2, eq3], (x, y, z))

print("Решения системы:")
if solutions:
    for sol in solutions:
        print(f"x = {sol[0]}, y = {sol[1]}, z = {sol[2]}")
else:
    print("Нет решений")
```

```python
from sympy import symbols, plot_implicit, Eq

x, y, z = symbols('x y z', real=True)

# Уравнения
eq1 = Eq(x - y + z, 2)
eq2 = Eq(x**2 + y**2 + z**2, 6)
eq3 = Eq(x**3 + y**3 + z**3, 8)

# Графики неявных функций (фиксируем z как параметр или константу, т.к. это 2D график)
# Например, зафиксируем z = 0 для построения сечения на плоскости z=0
z_value = 0

p1 = plot_implicit(eq1.subs(z, z_value), (x, -10, 10), (y, -10, 10),
                   line_color='blue',
                   show=False,
                   label="Плоскость: x - y + z = 2")

p2 = plot_implicit(eq2.subs(z, z_value), (x, -10, 10), (y, -10, 10),
                   line_color='red',
                   show=False,
                   label="Сфера: x² + y² + z² = 6")

p3 = plot_implicit(eq3.subs(z, z_value), (x, -10, 10), (y, -10, 10),
                   line_color='green',
                   show=False,
                   label="Кубическая поверхность: x³ + y³ + z³ = 8")

# Объединяем графики
p1.extend(p2)
p1.extend(p3)

# Настройки
p1.title = "Графики уравнений системы (сечение при z = 0)"
p1.xlabel = 'x'
p1.ylabel = 'y'
p1.legend = True

# Отображаем
p1.show()
```
